```
import pandas as pd
import numpy as np
```

```
hdata = pd.read_excel('/content/sample_data/XAU_Hdata.xlsx')
len = len(hdata)
vals = np.zeros(len, dtype='uint8')
```

```
X = np.array(hdata.HIGH)
```

```
test_level = 1
for i in range(test_level ,len-test_level):
  if X[i]>X[i-1] and X[i]>X[i+1]:
    vals[i]=1
```

```
hdata['TD_SUPPLY_LVL']=vals
```
